```python
import os

# Chemin du dossier contenant les fichiers
folder_path = "./All_Variables"

# Parcourir tous les fichiers dans le dossier
for filename in os.listdir(folder_path):
    # Vérifier si le fichier contient des espaces
    if " " in filename:
        # Nouveau nom avec les espaces remplacés par des underscores
        new_filename = filename.replace(" ", "_")
        # Chemins complets pour le renommage
        old_file = os.path.join(folder_path, filename)
        new_file = os.path.join(folder_path, new_filename)
        # Renommer le fichier
        os.rename(old_file, new_file)
        print(f"Renommé : {filename} -> {new_filename}")

print("Renommage terminé.")
```

```python
def process_dataframe(df):
    # Supprimer la colonne 'Unnamed: 673' si elle existe
    if 'Unnamed: 673' in df.columns:
        df = df.drop(columns=['Unnamed: 673'])
    
    # Créer un dictionnaire pour stocker les dataframes divisés
    dataframes = {}
    
    # Identifier les colonnes correspondant à chaque CPI
    for i in range(1, 17):  # CPI01 à CPI16
        cpi_columns = [col for col in df.columns if f'CPI{i:02}' in col]
        cpi_columns.insert(0, 'FechaHora')  # Ajouter la colonne 'FechaHora'
        
        # Créer un dataframe pour chaque CPI
        dataframes[f'CPI{i:02}'] = df[cpi_columns]
    
    return dataframes
```

```python
dataframes = process_dataframe(df)
```

```python
cpi01_df = dataframes['CPI01']
cpi01_df.head()
```

```python
def rename_columns(df):
    # Dictionnaire de correspondance pour renommer les colonnes
    rename_dict = {
        'CoilTemperature': 'Coil_Temperature',
        'CommsStatus': 'Communication_Status',
        'CosineOfPhi': 'Cosine_Phi',
        'DailyACInjectedEnergy': 'Daily_AC_Energy_Injected',
        'GridFrequency': 'Grid_Frequency',
        'GridRMSVoltageOfPhase1': 'Grid_RMS_Voltage_Phase1',
        'GridRMSVoltageOfPhase2': 'Grid_RMS_Voltage_Phase2',
        'GridRMSVoltageOfPhase3': 'Grid_RMS_Voltage_Phase3',
        'HeatSinkRTemperature': 'HeatSink_R_Temperature',
        'HeatSinkSTemperature': 'HeatSink_S_Temperature',
        'HeatSinkTTemperature': 'HeatSink_T_Temperature',
        'OutputActivePower': 'Output_Active_Power',
        'OutputGridRMSCurrentOfPhase1': 'Output_Grid_RMS_Current_Phase1',
        'OutputGridRMSCurrentOfPhase2': 'Output_Grid_RMS_Current_Phase2',
        'OutputGridRMSCurrentOfPhase3': 'Output_Grid_RMS_Current_Phase3',
        'OutputReactivePower': 'Output_Reactive_Power',
        'PVCurrent': 'PV_Current',
        'PVVoltage': 'PV_Voltage',
        'SignOfPhi': 'Sign_Phi',
        'TotalACEnergyInjected': 'Total_AC_Energy_Injected',
        'TotalHoursofOperation': 'Total_Hours_Operation'
    }
    
    # Renommer les colonnes en supprimant les préfixes et suffixes, puis en utilisant le dictionnaire
    new_columns = []
    for col in df.columns:
        # Supprimer les préfixes comme "ROOT.CPI01."
        if 'ROOT.CPI' in col:
            col = '.'.join(col.split('.')[2:])  # Supprime les deux premiers segments
        # Supprimer le suffixe ".hf"
        if col.endswith('.hf'):
            col = col.replace('.hf', '')
        # Remplacer les noms en utilisant le dictionnaire
        for key, value in rename_dict.items():
            if key in col:
                col = col.replace(key, value)
        new_columns.append(col)
    
    df.columns = new_columns
    return df
```

```python
cpi01_df = rename_columns(cpi01_df)
cpi01_df.head()
```
